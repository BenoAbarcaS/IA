**(0,5 pts) ¿Tiene sentido hacer un diagrama de torta para mostrar la cantidad de cambios que se han hecho por cargo?
Justifique su respuesta y en caso afirmativo realıcelo y discuta los resultados.**

El Pie Chart podria arrojar la cantidad de cambios, pero no seria el uso más adecuado para este, si quisieramos ver la cantidad de cambios, por ejemplo usaria un countplot. Si es que usase el Pie Chart seria como lo mostrado abajo, es decir, para ver cuales son los cargos más volatiles a cambiar de mando.

```python
index=data['position'].value_counts().index
values=data['position'].value_counts().values
```

```python
fig = plt.figure(figsize =(11, 11))
plt.pie(values,autopct='%1.1f%%')
labels = ['%s, %1.1f %%' % (l, s) for l, s in zip(index, values)]
fig.legend(labels,title="Position",loc="center",bbox_to_anchor=(1, 0, 0.5, 1))
```
